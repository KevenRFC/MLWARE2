### 2.2 Reemplazar valores

```python
data2 = data.copy()
```

```python
cont = 0
for y in vars_revision_group:
    cont = cont + 1
    print("------------------------------------------------------")
    print(str(cont),". Var: ",y, sep = "")
    print(pd.concat([pd.DataFrame(base[:][y].value_counts().index, columns = ['Atributo']),
           pd.DataFrame(base[:][y].value_counts().values, columns = ['Cantidad']),
           pd.DataFrame(100*base[:][y].value_counts().values/len(base), columns = ['%Total'])], axis = 1))
```

```python
# PERFIL_TACTICO
feature = 'PERFIL_TACTICO'
data2.loc[data2[feature] == 'CONSUMO JOVEN', feature] = 'CONS_JOV'
data2.loc[data2[feature] == 'CONSUMO', feature] = 'CONSUMO'
data2.loc[data2[feature] == 'JÓVENES ALTO POTENCIAL', feature] = 'JOV_ALTO_POT'
data2.loc[data2[feature] == 'TOP CONSUMO', feature] = 'TOP_CONS'
data2.loc[data2[feature] == 'EXCLUSIVO ALTO POTENCIAL', feature] = 'EXCLUS_ALTO_POT'
data2.loc[data2[feature] == 'EXCLUSIVO', feature] = 'EXCLUSIVO'
data2.loc[data2[feature] == 'SIN INFO', feature] = 'SIN_INFO'
data2.loc[data2[feature] == 'BAJOS INGRESOS', feature] = 'BAJOS_INGR'
data2.loc[data2[feature] == 'ENALTA', feature] = 'ENALTA'
data2.loc[(data2[feature] != 'CONS_JOV') &
          (data2[feature] != 'CONSUMO') &
          (data2[feature] != 'JOV_ALTO_POT') &
          (data2[feature] != 'TOP_CONS') &
          (data2[feature] != 'EXCLUS_ALTO_POT') &
          (data2[feature] != 'EXCLUSIVO') &
          (data2[feature] != 'SIN_INFO') &
          (data2[feature] != 'BAJOS_INGR') &
          (data2[feature] != 'ENALTA'), feature] = 'OTROS'
```

```python
# TIPESTCIVIL
feature = 'TIPESTCIVIL'
data2.loc[(data2[feature] != 'SOL') &
          (data2[feature] != 'CAS'), feature] = 'OTROS'
```

```python
# TIPSITUACIONLABORAL
feature = 'TIPSITUACIONLABORAL'
data2.loc[(data2[feature] != 'DEP') &
          (data2[feature] != 'SUC') &
          (data2[feature] != 'IND') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'EST'), feature] = 'OTROS'
```

```python
# TIPNIVELEDUCACIONAL
feature = 'TIPNIVELEDUCACIONAL'
data2.loc[(data2[feature] != 'BAC') &
          (data2[feature] != 'TEC') &
          (data2[feature] != 'TIT') &
          (data2[feature] != 'SUP') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'TEI') &
          (data2[feature] != 'MAE'), feature] = 'OTROS'
```

```python
# EDAD
feature = 'EDAD'
data2.loc[data2[feature] <= 21, feature] = 21
data2.loc[data2[feature] >= 62, feature] = 62
```

```python
# CODUBIGEO
feature = 'CODUBIGEO'
data2.loc[(data2[feature] != 150140) &
          (data2[feature] != 150122) &
          (data2[feature] != 150114) &
          (data2[feature] != 150135) &
          (data2[feature] != 150130) &
          (data2[feature] != 150132) &
          (data2[feature] != 150101) &
          (data2[feature] != 150131) &
          (data2[feature] != 150136) &
          (data2[feature] != 150117) &
          (data2[feature] != 150103) &
          (data2[feature] != 70101) &
          (data2[feature] != 150108) &
          (data2[feature] != 150110) &
          (data2[feature] != 130101) &
          (data2[feature] != 150133) &
          (data2[feature] != 150141) &
          (data2[feature] != 150121) &
          (data2[feature] != 200101), feature] = 999999
```

```python
# DEPARTAMENTO
feature = 'DEPARTAMENTO'
data2[feature] = data2[feature].map(lambda x: x.replace(' ', ''))
data2.loc[data2[feature] == 'LA LIBERTAD', feature] = 'LA_LIBERTAD'
data2.loc[(data2[feature] != 'DEP') &
          (data2[feature] != 'SUC') &
          (data2[feature] != 'IND') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'NDI') &
          (data2[feature] != 'EST'), feature] = 'OTROS'
```

```python
# FLG_ACT_PAS
feature = 'FLG_ACT_PAS'
data2.loc[data2[feature] == 'A/P', feature] = 'AP'
data2.loc[data2[feature] == 'P', feature] = 'P'
data2.loc[(data2[feature] != 'AP') &
          (data2[feature] != 'P'), feature] = 'OTROS'
```

```python
# FLG_ACT_PAS_SALDO
feature = 'FLG_ACT_PAS_SALDO'
data2.loc[data2[feature] == 'A/P', feature] = 'AP'
data2.loc[data2[feature] == 'P', feature] = 'P'
data2.loc[(data2[feature] != 'AP') &
          (data2[feature] != 'P'), feature] = 'OTROS'
```

```python
cont = 0
for y in vars_revision_group:
    cont = cont + 1
    print("------------------------------------------------------")
    print(str(cont),". Var: ",y, sep = "")
    print(pd.concat([pd.DataFrame(data2[:][y].value_counts().index, columns = ['Atributo']),
           pd.DataFrame(data2[:][y].value_counts().values, columns = ['Cantidad']),
           pd.DataFrame(100*data2[:][y].value_counts().values/len(data2), columns = ['%Total'])], axis = 1))
```
